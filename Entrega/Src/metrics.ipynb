```python
import pandas as pd
import ast
```

```python
df = pd.read_csv('../data/mixtral_inferred_candidates.csv')
df.head()
```

```python
def add_dictionaries(dict_list : list[dict]):
    """
    Add dictionaries in a list together
    """
    master_dict = {}
    
    for d in dict_list:
        d = ast.literal_eval(d)
        for k, v in d.items():
            if k in master_dict:
                master_dict[k] += v
            else:
                master_dict[k] = v
                
    # sort dictionary by values
    master_dict = dict(sorted(master_dict.items(), key=lambda item: item[1], reverse=True))
    
    return master_dict

df['Inferred_candidates'] = None

for i, row in df.iterrows():
    inferred_candidates = [row['Chatgpt_inferred_candidates'], row['Mixtral_inferred_candidates'], row['Gemini_inferred_candidates'], row['Llama_inferred_candidates'], row['Palm2_inferred_candidates']]
    aux = add_dictionaries(inferred_candidates)
    df.at[i, 'Inferred_candidates'] = aux

```

```python
def dict_to_list(d : dict):
    """
    Convert a dictionary to a string list of its values separated by \t
    """
    return '\t'.join([f'{k}' for k, v in d.items()])
df['Inferred_candidate_list'] = df['Inferred_candidates'].apply(dict_to_list)
df.head()
```

```python
output_df = df[['Text','Complex_word','Inferred_candidate_list']]
output_df.to_csv('../data/inferred_candidates.tsv', index=False, sep='\t', header=False, encoding='utf-8')
```
