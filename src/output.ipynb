```python
import pandas as pd
```

```python
df = pd.read_csv('../data/total_inferred_candidates.csv')
df.head()
```

```python
def parse_row(row):
    line = ''
    
    line += row['Text'] +'\t'
    line+= row['Complex_word']
    
    candidate_dict = eval(row['Total_inferred_candidates'])
    
    for key in candidate_dict:
        line += '\t' + key
    
    return line


with open('../data/total_inferred_candidates.tsv', 'w', encoding='utf-8') as f:
    for index, row in df.iterrows():
        f.write(parse_row(row) + '\n')
        
f.close()
```
